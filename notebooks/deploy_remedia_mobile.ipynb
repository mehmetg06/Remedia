# Remedia — Telefonda Tek Seferlik Deploy

Bu notebook yalnızca ilk kurulum içindir. Aşağıdaki hücreyi bir kez çalıştır.

İşlem sonunda sabit Modal adresi ve Jupyter giriş token'ı gösterilir. Token'ı güvenli bir yere kaydet.


```python
import re
import secrets
import subprocess
import sys
import urllib.request
from pathlib import Path

subprocess.run([sys.executable, '-m', 'pip', 'install', '--quiet', '--upgrade', 'modal'], check=True)
token = secrets.token_urlsafe(24)
subprocess.run([sys.executable, '-m', 'modal', 'secret', 'create', 'remedia-jupyter-auth', f'JUPYTER_TOKEN={token}', '--force'], check=True)
script = Path('/tmp/remedia_web_jupyter.py')
urllib.request.urlretrieve('https://raw.githubusercontent.com/mehmetg06/Remedia/main/modal/remedia_web_jupyter.py', script)
proc = subprocess.run([sys.executable, '-m', 'modal', 'deploy', str(script)], text=True, capture_output=True)
print(proc.stdout)
if proc.returncode != 0:
    print(proc.stderr)
    raise RuntimeError('Deploy başarısız oldu.')
urls = re.findall(r'https://[^\s]+\.modal\.run', proc.stdout + '\n' + proc.stderr)
print('\n✅ Deploy tamamlandı')
if urls:
    print('Açılacak sabit adres:', urls[-1])
else:
    print('Sabit adres Modal Apps ekranında remedia-web-jupyter altında görünecek.')
print('Jupyter token:', token)
print('Bu tokenı Notlar uygulamasına kaydet.')
```
