# Remedia — Telefonda Tek Seferlik Deploy

Bu notebook yalnızca ilk kurulum ve güncellemeler içindir. Aşağıdaki hücreyi çalıştır.

Erişim ayrı Jupyter token'ı yerine Modal hesabı oturumuyla korunur.


```python
import re
import subprocess
import sys
import urllib.request
from pathlib import Path

subprocess.run([sys.executable, '-m', 'pip', 'install', '--quiet', '--upgrade', 'modal'], check=True)
script = Path('/tmp/remedia_web_jupyter.py')
urllib.request.urlretrieve('https://raw.githubusercontent.com/mehmetg06/Remedia/main/modal/remedia_web_jupyter.py', script)
proc = subprocess.run([sys.executable, '-m', 'modal', 'deploy', str(script)], text=True, capture_output=True)
print(proc.stdout)
if proc.returncode != 0:
    print(proc.stderr)
    raise RuntimeError('Deploy başarısız oldu.')
urls = re.findall(r'https://[^\s]+\.modal\.run', proc.stdout + '\n' + proc.stderr)
print('\n✅ Deploy tamamlandı')
if urls:
    base = urls[-1].rstrip('/')
    print('Aç:', base + '/lab/tree/Remedia/notebooks/remedia_modal.ipynb')
else:
    print('Adres Modal Apps ekranında remedia-web-jupyter altında görünecek.')
print('Ayrı Jupyter tokenı gerekmiyor; Modal hesabınla giriş yap.')
```
