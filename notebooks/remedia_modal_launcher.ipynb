# Remedia — Kalıcı Modal Launcher

Bu notebook'u yalnızca bir kez import et. Bundan sonra aynı notebook'u açıp aşağıdaki tek hücreyi çalıştır.

Hücre her çalıştığında GitHub'daki en güncel Remedia kurulumunu ve formunu çeker. Yeni sürüm çıktığında yeniden notebook oluşturman veya import etmen gerekmez.


```python
import time
import urllib.request

BOOTSTRAP_URL = (
    'https://raw.githubusercontent.com/mehmetg06/Remedia/'
    'main/scripts/remedia_modal_bootstrap.py'
)
request = urllib.request.Request(
    f'{BOOTSTRAP_URL}?t={time.time_ns()}',
    headers={'Cache-Control': 'no-cache', 'User-Agent': 'Remedia-Modal-Launcher'},
)
with urllib.request.urlopen(request, timeout=60) as response:
    bootstrap_source = response.read().decode('utf-8')

exec(compile(bootstrap_source, BOOTSTRAP_URL, 'exec'), globals())
run(globals())
```
